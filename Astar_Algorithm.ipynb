```python
#Creat a class to set up the problem
class Problem(object):
#defining the heuristic values for each node    
    def heuristic(self, node, goal):
        heuristic = {
            'A': 4,
            'B': 5,
            'C': 6,
            'D': 4,
            'E': 3,
            'F': 2,
            'H': 3,
            'G1': 0,
            'G2': 0,
            'S': 6,
        }
        
        return heuristic[node]

#defining connecting nodes in the problem   
    def neighbour_nodes(self, node):
        if node == 'A':
            return['G1', 'B']
        if node == 'B':
            return['A', 'C']
        if node == 'C':
            return['S', 'F']
        if node == 'D':
            return['E', 'C']
        if node == 'E':
            return['G2', 'S']
        if node == 'F':
            return['G1', 'B']
        if node == 'H':
            return['D', 'G2']
        if node == 'S':
            return['A', 'B', 'D']

#defining the path cost  between the connecting nodes in problem
    def g_score_between_node(self, node2, node1):
        if node1 == 'A' and node2 == 'B':
            return 3
        if node1 == 'A' and node2 == 'G1':
            return 9
        if node1 == 'B' and node2 == 'A':
            return 5
        if node1 == 'B' and node2 == 'C':
            return 3
        if node1 == 'C' and node2 == 'S':
            return 4
        if node1 == 'C' and node2 == 'F':
            return 5
        if node1 == 'D' and node2 == 'C':
            return 6
        if node1 == 'D' and node2 == 'E':
            return 3
        if node1 == 'E' and node2 == 'S':
            return 10
        if node1 == 'E' and node2 == 'G2':
            return 4
        if node1 == 'F' and node2 == 'G1':
            return 8
        if node1 == 'F' and node2 == 'B':
            return 5
        if node1 == 'S' and node2 == 'A':
            return 8
        if node1 == 'S' and node2 == 'B':
            return 4
        if node1 == 'S' and node2 == 'D':
            return 6
        if node1 == 'H' and node2 == 'G2':
            return 6
        if node1 == 'H' and node2 == 'D':
            return 10
  
        
        
#defining the goal nodes
    def is_goal(self, node, goal):
        return node == 'G1' or node == 'G2'
        
```

```python
#create a function to find the path between initial node and goal node.
def find_path(problem, start, goal):
#begin with no nodes for the problem   
#explore - contains all connected nodes but not sorted in any order    
    explore = set()
#the agenda - contains neighbouring nodes, however it orders them with their f-score
    agenda = []
#visited - contains nodes that have been explored with the lowest f score from the agenda, nodes with a higher f score will not be added to the agenda
    visited = {}
#create a dictionary to store the g_score for each node
    g_score = {}
#create a dictionary to store the h_score for each node
    h_score = {}

#create a function to compute the f score every time
    def f_score(node):
        return g_score[node] + h_score[node]

#start the algorithm by placing the initial node to be explored   
    g_score[start] = 0
    h = problem.heuristic(start, goal)
    h_score[start] = h
    explore.add(start)
    agenda.append((f_score(start), start))

#use a while loop to search until a goal has been found or all paths have been explored
    while explore:
        agenda.sort()
        print('nodes place on agenda to visit are:', agenda)
        next_f, node = agenda.pop(0)
        explore.remove(node)
        if problem.is_goal(node, goal):
            path = [ node ]
            while node != start:
                node = visited[node]
                path.append(node)
            path.reverse()
            return path
        explore.add(node)
    
#adding the path score for connecting nodes
        for neighbour in problem.neighbour_nodes(node):
            if not neighbour in explore:
                new_g_score = g_score[node] + problem.g_score_between_node(neighbour, node)
 
                if neighbour not in explore:
#exploring new nodes and checking to see if there f score is lower than those a;ready in the agenda
                    visited[neighbour] = node
                    g = new_g_score
                    h = problem.heuristic(neighbour, goal)
                    g_score[neighbour] = g
                    h_score[neighbour] = h
                    explore.add(neighbour)
                    f = g + h
                    agenda.append( (f, neighbour) )
                else:
                    if new_g_score < g_score[neighbour]:
                        visited[neighbour] = node
                        g = new_g_score
                        g_score[neighbour] = g
                        h = problem.heuristic(neighbour, goal)
                        h_score[neighbour] = h
                        f = g + h
 
                        problem.on_update(neighbour, f, g, h)

solution = find_path(Problem(), 'S', ['G1', 'G2'])
print('\nThe path found by the A* algortihm = ', solution)
```
